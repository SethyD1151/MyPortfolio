Clean the Data.

```python
#Drop unnecessary columns.
df=nba_df.drop("Index", axis=1)
```

```python
 #Rename columns.
 df1=df.rename(columns={'GamesPlayed':'Games Played','RBs':'Rebounds','ASTs':'Assists','STLs':'Steals','BLKs':'Blocks','Ppg':'Points per Game','MinPlayed':'Minutes Played'})
```

```python
#Group data by Position for further analysis.
jkl=df1.groupby("Position")
```

```python
df_nba=df1.sort_values("Points per Game", ascending=False)#Sort values by point average.
nba=df_nba.head(5)#Dispay to 5 values.
#Visualize top scorers in a bar chart
gg=sns.barplot(nba, y="Player Name", x="Points per Game", orient='h')
gg.set_xlim(0,40)
gg.set_title('NBA Top Scorers 2022-2023')
gg.bar_label(gg.containers[0])
#Print Text
print("These are the top 3 scorers in the NBA as of last season.")
print("Joel Embiid was the top scorer in the 2022-2023 NBA season.")
print("If you look closely you will see he plays as a Center and scores more on average than point gaurds, and power forwards.")
nba
```
